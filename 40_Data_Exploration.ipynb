```python
country = 'India'

print(int(df[df['clean_country']== country ]['price'].min()))
print(int(df[df['clean_country']== country ]['price'].mean()))
print(int(df[df['clean_country']== country ]['price'].max()))
```

# 3 Data Analysis

```python
for country in df['clean_country'].unique():
    print(f"Country: {country}")
    print(f"Minimum Price: {int(df[df['clean_country']== country ]['price'].min())}")
    print(f"Mean Price: {int(df[df['clean_country']== country ]['price'].mean())}")
    print(f"Maximum Price: {int(df[df['clean_country']== country ]['price'].max())}")
    print("-" * 20)
```

# 4 Data Visualization

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

country_mean_prices = df.groupby('clean_country')['price'].mean().reset_index()


top_10_countries = country_mean_prices.sort_values(by='price', ascending=False).head(10)

plt.figure(figsize=(9, 6))
sns.barplot(x='clean_country', y='price', data=top_10_countries)
plt.xticks(rotation=45, ha='right')
plt.title('Top 10 Countries by Mean Price')
plt.tight_layout()
plt.show()
```
